```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)]
```

```python
def c_s(x):
    token=x.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None 
```

```python
df4=df3.copy()
df4['total_sqft']=df4['total_sqft'].apply(c_s)
df4.head(50)
```

```python
df5=df4.copy()
df5['price_per_sqft']=df5['price']*100000/df5['total_sqft']
df5.head()
```

```python
len(df5.location.unique())
```

```python
df5.location=df5.location.apply(lambda x: x.strip())
loc_stats=df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
loc_stats
```

```python
other_loc=loc_stats[loc_stats<=10]
other_loc
```

```python
df5.location=df5.location.apply(lambda x: 'other' if x in other_loc else x)
len(df5.location.unique())
```

```python
df6=df5[~(df5.total_sqft/df5.bhk<300)]
df6.shape
```

```python
def remove_outliers(df):
    df_out=pd.DataFrame()
    for a,sub_df in df.groupby('location'):
        m=np.mean(sub_df.price_per_sqft)
        st=np.std(sub_df.price_per_sqft)
        reduced_df=sub_df[(sub_df.price_per_sqft>(m-st))&(sub_df.price_per_sqft<(m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7=remove_outliers(df6)
df7.shape
```

```python
def psc(df,location):
    bhk2=df[(df.location==location) &(df.bhk==2)]
    bhk3=df[(df.location==location) &(df.bhk==3)]
    matplotlib.rcParams['figure.figsize']=(15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='BHK2',s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,color='green',label='BHK3',s=50)
    plt.xlabel("Total Square Feet")
    plt.ylabel("Price ")
    plt.title(location)
    plt.show()
    
```

```python
psc(df7,"Hebbal")
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')



df8 = remove_bhk_outliers(df7)
df8.shape
```

```python

df9 = df8[df8.bath<df8.bhk+2]
df9.shape
```

```python
df9.head(2)
```

```python
df10 = df9.drop(['size','price_per_sqft'],axis='columns')
df10.head(3)
```

```python
dummies = pd.get_dummies(df10.location)
dummies.head(3)
```

```python
df11 = pd.concat([df10,dummies.drop('other',axis='columns')],axis='columns')
df11.head()
```

```python
df12 = df11.drop('location',axis='columns')
df12.head(2)
```

```python
df12.shape
```

```python
X = df12.drop(['price'],axis='columns')
X.head(3)
```

```python
y = df12.price
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1000, 2, 2)
```
